## 使用正则表达式检测词语模式

```python
import re
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
```

### 使用基本元字符

函数re.search(p,s)检查字符串s中是否有模式p

使用正则表达式“ed$”查找以ed结尾的词汇

```python
[w for w in wordlist if re.search('ed$',w)]
```

## 元字符

通配符“.”匹配任何字符， 通配符“?”表示前面的字符可
选， “^”匹配字符串的开始，“$”匹配字符串的结尾

```python
[w for w in wordlist if re.search('^..j..t..$',w)]
```

方括号[]：范围限制。其中的字符没有顺序

T9输入法。 有哪些词汇由序列4653产生？

```python
[w for w in wordlist if re.search('^[ghi][mno][jlk][def]$',w)]
```

‘-’表示区间，如«^[g-o]+$»

'^'单独使用表示开头，‘^’ 在方括号内的第一个字符位置时出现表示排除

例如： «[^aeiouAEIOU]»匹配除元音字母之外的所有字母

### 闭包

“+”表示的是前面的项目的一个或多个实例

“*” 表示前面的项目的零个或多个实例

```python
chat_words = sorted(set(w for w in nltk.corpus.nps_chat.words()))
[w for w in chat_words if re.search('^o+h+$', w)]
```

‘\’表示其后面的字母不再有特殊的含义而是按字面匹配
    
‘{}’指定前面项目的重复次数

‘|’ 字符表示从其左边的内容和右边的内容中选择一个

‘()’表示操作符的范围，可以与管道（‘|’ ） 符号一起使用

#### 匹配小数

```python
wsj = sorted(set(nltk.corpus.treebank.words()))#华尔街日报
[ w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]
```

#### 匹配货币

```python
[w for w in wsj if re.search('^[A-Z]+\$$', w)]
```

#### 匹配年份

```python
[w for w in wsj if re.search('^[0-9]{4}$',w)]
```

#### 匹配单词

```python
[w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}', w)]#第一个单词超过5个字母，第三个单词小于6个字母
```

```python
[w for w in wsj if re.search('(ed|ing)$',w)] #|前后不要加空格
```

### 提取字符块

```python
#找出元音
word='ahskiosuulfljdshflb'
re.findall(r'[aeiou]',word)
```

找出文本中的两个或两个以上的元音序列，并确定它们的相对频率

```python
fd= nltk.FreqDist(vs for word in wsj
                     for vs in re.findall(r'[aeiou]{2,}', word))
fd.items()
```

忽略词内部的元音（首尾则保留）

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in udhr[:75]))
```
